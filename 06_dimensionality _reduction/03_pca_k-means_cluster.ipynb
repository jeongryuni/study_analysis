## PCA를 통해 차원이 많은 데이터 축소 후 k-means 클러스터링 하여 데이터가 가진 특성을 파악하는 분석

## 데이터 불러오기

```python
import pandas as pd

credit_df1 = pd.read_csv('data/CC_GENERAL.csv')
credit_df1.head(10)
```

## 데이터 표준화

```python
# 아이디를 CUST_ID를 제외한 고객 정보 분리
credit_df2 = credit_df1.drop(columns=['CUST_ID'], axis=1)
credit_df2.describe() # 기초 통계량 확인
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_credit_df2 = scaler.fit_transform(credit_df2)
scaled_credit_df2
```

```python
# PCA
from sklearn.decomposition import PCA

# 객체
X = scaled_credit_df2.copy()

# 적용
pca = PCA(n_components=2) # 2차원으로 축소
x_pca = pca.fit_transform(scaled_credit_df2)
pca_df = pd.DataFrame(x_pca)
pca_df.head()
```

## K-means 클러스터링

```python
# Elbow plt
from sklearn.cluster import KMeans
import seaborn as sns

ks = range(1, 10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(pca_df)
    inertias.append(model.inertia_)

# 시각화
sns.lineplot(x=ks, y=inertias, marker='o')
```

k의 값이 6일 때부터 inertia 값의 감소폭이 작아짐으로 k를 6으로 지정한 후에 k-means 실행

```python
model = KMeans(n_clusters=6)
model.fit(pca_df)
labels = model.labels_ # 고객마다 부여할 클러스터
x, y = pca_df[0], pca_df[1]

# PCA 진행한 데이터프레임에 클러스터 번호 부여
pca_km_df = pd.DataFrame({'PC1':x, 'PC2':y, 'label':labels})

# 시각화
sns.scatterplot(data=pca_km_df, x = 'PC1', y = 'PC2', hue = 'label', palette='rainbow')
```

클러스터 3과 4에 속한 고객들이 가장 많이 흩어져 있으며 나머지 클러스터에 속한 고객들은 상대적으로 유사한 특징

## 클러스터 결과 해석

```python
# k = 6인 k-means 클러스터링 진행
model = KMeans(n_clusters=6, random_state=111)
model.fit(pca_df)

labels = model.labels_ # 고객마다 부여할 클러스터

# 원본 데이터에 클러스터 부여
credit_df1['cluster'] = labels
```

```python
# 부여한 클러스터 번호를 토대로, 각 클러스터별 할당된 고객의 숫자를 확인
credit_df1['CUST_ID'].groupby(credit_df1['cluster']).count()
```

클러스터 1이 가장 많고 클러스터 5에 속한 고객이 제일 적다.

```python
# 클러스터에 따라 각 열의 값 차이
# 열이 많기 때문에 편의를 위해 행과 열을 바꿔서 출력
credit_df1.groupby(credit_df1['cluster']).mean(numeric_only=True).T
```

여러 컬럼들 중 대출과 관련 있어 보이는 잔액(BALANCE), 잔액 갱신 빈도(BALANCE_FREQUENCY)
두 개의 컬럼을 활용하여 특징 살펴보기

```python
credit_df1.groupby(credit_df1['cluster']).mean(numeric_only=True).T.iloc[0:2,:]
```

위 분석을 토대로 대출 서비스를 제안한다고 가정했을 때 평균 잔액이 적고 잔액 갱신 빈도가 낮아 대출에 대한 수요가 있을 것으로 
추정되는 클러스터 1과 클러스터 4의 고객들에게 제안하는 것이 합리적인 제안이라고 판단
