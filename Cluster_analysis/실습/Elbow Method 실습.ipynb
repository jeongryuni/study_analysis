## 실습 설명
주어진 데이터를 활용해 k-means의 inertia 값을 구할 수 있는 그래프를 그려주세요.

Elbow Method를 그리기 위한 클러스터 개수는 1부터 15까지로, random_state는 123으로 설정해 주세요.

```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
%matplotlib inline
```

```python
scaled_df = pd.read_csv('data/scaled_data.csv', index_col=[0])
inertias = []

for k in range(1, 16):
    model = KMeans(n_clusters=k, random_state=123)
    model.fit(scaled_df)
    inertias.append(model.inertia_)

# 시각화
sns.set(style="darkgrid")

sns.lineplot(x=range(1,16), y=inertias, marker='o')
```
