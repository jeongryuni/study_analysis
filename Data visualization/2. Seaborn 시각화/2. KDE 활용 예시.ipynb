### 2. 박스플롯

```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('data/body.csv', index_col=0)
body_df.head()
```

```python
body_df.plot(kind='hist', y='Height', bins=15)
```

```python
# KDE 활용
# 1. Seaborn을 사용하면 히스토그램 위에 Seaborn을 얹을 수 있다.
sns.distplot(body_df['Height'], bins=15)
```

```python
# 박스플롯
body_df.plot(kind='box', y='Height')
```

## 바이올린 플롯
-- KDE그래프를 대칭되게 펼친 모양
-- 분포 전체를 보여줌
-- 시각적으로 근사함

```python
sns.violinplot(y=body_df['Height'])
```

### 키와 몸무게 연관성
#### -- 산점도

```python
body_df.plot(kind='scatter', x='Height', y='Weight')
```

전반적으로 키가 클수록 체중이 같이 올라간다.

## 등고선
#### -- x축에는 키에 대한 KDE
#### -- y축에는 체중에 대한 KDE
#### -- 두 개를 합쳐서 3차원으로 만듬
#### -- 연관성과 분포를 같이 살펴볼 수 있다.

## 선이 가까울수록 가파르고
## 선이 멀수록 평평하다.

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

```python
# 키는 평평하다가 가파르게 오르고 가파르게 떨어진다.
sns.kdeplot(body_df['Height'])
```

```python
# 체중은 위에서 평평해진다.
sns.kdeplot(body_df['Weight'])
```
