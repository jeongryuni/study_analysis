이번에도 예산이 너무 큰 영화 몇 개를 제거해보려고 합니다.

하지만 이번에는 IQR이 아니라 예산 상위 15개를 제거하려고 하는데요.

movie_metadata.csv에서 예산이 가장 높은 15개 영화를 제거하고, 산점도를 그려보세요.

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/movie_metadata.csv')
df
```

```python
df.plot(kind='scatter', x='budget', y='imdb_score')
```

```python
budget_head = df['budget'].sort_values(ascending=False).head(15)
budget_head.index
```

```python
df.drop(budget_head.index, inplace=True)
df
```

```python
df.plot(kind='scatter', x='budget', y='imdb_score')
```
