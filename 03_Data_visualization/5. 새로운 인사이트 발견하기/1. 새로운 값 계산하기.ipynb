```python
# 새로운 Total 컬럼에 컬럼들의 합계 넣기
df['Total'] = df.sum(axis='columns')
df
```

```python
df.plot(y='Total')
```

전체 시청률은 떨어지고 있다.

```python
# Group 1 = 지상파의 각 해의 시청률 더하기
# Group 2 = 그 외의 종편 시청률 더하기
df['Group 1'] = df.loc[:, 'KBS':'SBS'].sum(axis='columns')
df['Group 2'] = df.loc[:, 'TV CHOSUN':'MBN'].sum(axis='columns')
```

```python
df
```

```python
df.plot(y=['Group 1', 'Group 2'])
```

전체 시청률은 줄어드는데, 종편은 시청률이 오르고 있다.
